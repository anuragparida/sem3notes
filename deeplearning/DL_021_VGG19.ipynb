## Using a pretrained modell VGG19



```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from  tensorflow.keras.applications.vgg19 import preprocess_input, VGG19, decode_predictions
import matplotlib.pyplot as plt 
```

```python
image = load_img("oldtimer.jpg", target_size=(224, 224))
image = img_to_array(image)
image = image.reshape((1, 224, 224, 3))
plt.imshow(image[0]/255)

#now preprocess like it was done during training of CGG (rescale to -1 to 1)
image = preprocess_input(image)
```

```python
model = VGG19()
pred = model.predict(image)
```

```python
decode_predictions(pred)
```
