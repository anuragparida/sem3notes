# Introducing the Exponentially Weighted Moving Average

```python
# plotting a noisy curve 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,150)
y = ((-(x-4)**2)+35)+np.random.randn(len(x))*7


plt.plot(x,y)
```

## Denoise with the Simple Moving Average  

```python
import pandas as pd
ser = pd.Series(y)
z = ser.rolling(window=3).mean()
plt.plot(x,y)
plt.plot(x,z)
```

## Denoise with the Exponentially Weighted Moving Average

```python
ewm = ser.ewm(alpha=0.1).mean()
plt.plot(x,y)
plt.plot(x,ewm)
```
